Question 1  - Breast density Risk factor

```r
#Loading in data, seperated out those whoe developed cancer from those who did not on excel
cancer<- read.csv("data2.csv")
control<- read.csv("data1.csv")


#Small check to make sure the data is in the right place
head(cancer)
head(control) 
```

```r
# Summary data of patients who developed cancer 
summary(cancer)
```

```r
# Summary data of patients who did not develop cancer 
summary(control)
```

```r
#loading the compelte data set

total<- read.csv("total.csv")

head(total)
```

```r
#Boxplot comparing the mean breast density of the two groups.

# Load ggplot2
library(ggplot2)
 
 
# A boxplot.
ggplot(total, aes(x=as.factor(case), y=density)) + 
    geom_boxplot(fill="blue1", alpha=0.2) + 
    xlab("Group")
```

```r
#Same Boxplot as above, but in the consistent colour scheme and graphical style

boxplot(control$Density, cancer$Density, col= c(rep("red", 1),rep("blue",1)), xlab = "group", ylab = "Breast Density", main = "Breast Density between those developed cancer and those who dont")
```

```r
# Mean and Standard deviation breast density of patients who developed cancer 
apply(cancer[5], 2, mean)
apply(cancer[5], 2, sd)

# Mean and Standard deviation breast density of who didnt developed cancer 
apply(control[5], 2, mean)
apply(control[5], 2, sd)

#Number of Patinets in each Cohort
nrow(cancer)
nrow(control)
```

```r
#T-Test comparing the mean breast density between patients who developed cancer and those who didnt

t.test(cancer[5],control[5])
```
